```python
import numpy as np
from pylab import plot,xlabel,ylabel,show

# Constants
L = 20     # Thickness of steel in meters
D = .1   # Thermal diffusivity
N = 100       # Number of divisions in grid
a = L/N       # Grid spacing
h = 1e-4      # Time-step
epsilon = h/1000

Tstart = 10    # Initial temperature in Celcius

t1 = 9
t2 = 9.25
t3 = 9.5
t4 = 9.75
tend = t4 + epsilon

# Create arrays
T = np.empty(N+1,float)
T[0:N] = Tstart
Tp = np.empty(N+1,float)

# Main loop
t = 0.0
c = h*D/(a*a)
while t<tend:
    # Calculate the new values of T
    T[1:N] += 10 + 12*np.sin((2*np.pi*t)/365)
    for i in range(1,N):
        Tp[i] = T[i] + c*(T[i+1]+T[i-1]-2*T[i])
    T,Tp = Tp,T
    t += h

    # Make plots at the given times
    if abs(t-t1)<epsilon:
        plot(T)
    if abs(t-t2)<epsilon:
        plot(T)
    if abs(t-t3)<epsilon:
        plot(T)
    if abs(t-t4)<epsilon:
        plot(T)

xlabel("x")
ylabel("T")
show()
```
